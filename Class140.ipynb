```
import pandas as pd 
import numpy as np 

df1=pd.read_csv('tmdb_5000_credits.csv')
df2=pd.read_csv('tmdb_5000_movies.csv')
```

# Demographic Filtering

```
C = df2['vote_average'].mean()
print(C)
```

```
m = df2['vote_count'].quantile(0.9)
print(m)
```

```
q_movies = df2.copy().loc[df2['vote_count'] >= m]
print(q_movies.shape)
```

```
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)

q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```
q_movies = q_movies.sort_values('score', ascending=False)
q_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

```
import plotly.express as px

fig = px.bar((q_movies.head(10).sort_values('score', ascending=True)), x="score", y="title", orientation='h')
fig.show()
```

```
from ast import literal_eval

features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(literal_eval)

df2.dtypes
```

Here it looks like all the data in the list of dictionaries. Howerver there might be few rows which are in string format.

To ensure All rows are in list of dictionaries, we use a python module, literal_eval.

```
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan

df2['director'] = df2['crew'].apply(get_director)
```

We want to find the name of director whic is available in the crew column.
We weill create a seperate director column.

It will written NaN if there is no name.

isinstance cross checks if the value of the column is a list or not. 

The function get_list checks if the value x is an instance of a list or not.

If it is not then return an empty list.

If it a list of dictionaries the find all the values in all the dictionaries inside it and store them in a list.

```
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        return names
    return []

features = ['cast', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

COnvert to lower case, remove spaces from the elements

Cast, Keyword, Genre are a list and director is a string

clean_data checks is the value is a list using isinstance function.

convert to lower case, replace(" ") with("")

if NaN return empty string

apply clean_data function to all the columns cast, keywords, director, genre

```
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''

features = ['cast', 'keywords', 'director', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

create_soup function joins with space between columns as we are not applying the functions for a particular function but to an entire row of a dataframe by giving attribute axis =1.

join function takes all the elements of the list and joins them together and returns a new string.

```
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df2['soup'] = df2.apply(create_soup, axis=1)
```

countVectorizer counts the words after removing the stop_words then we are  coverting into a 3d matrix or 3d array.

Stop words are the words that don't add any value to a given sentence but are there to make it gramatically correct.
eg - and, but, the etc.

```
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df2['soup'])

print(count)
```

cosine_similarity - it is a classifier

```
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

we want to check the index of our movie data to the movie title and also changing the index of title.

create a function using cosine_similarity classifier

we are passing the title of the movie that the user like to our classifier.

```
df2 = df2.reset_index()
indices = pd.Series(df2.index, index=df2['title'])
```

We are then finding the index of the movie in our dataframe using the indices variable we created earlier, which contains the indexes of all the movies in the dataframe. We created this when we changed the index of our dataframe to the title of the movie. Next, we are creating a list of all the scores of the movies. This is the score of similarity of each movie with what the user likes. We are then using the sorted function on our data to sort the scores of all the movies

```
def get_recommendations(title, cosine_sim):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return df2['title'].iloc[movie_indices]
```
